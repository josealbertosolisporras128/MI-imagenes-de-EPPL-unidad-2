### 📝REDACCIÓN DEL PROBLEMA: Sistema de Cálculo Geométrico Interactivo
Planteamiento del Problema:
En el estudio de la geometría, es fundamental comprender y aplicar las fórmulas para calcular las propiedades de las figuras circulares.
Los estudiantes de nivel preparatoria frecuentemente necesitan realizar cálculos del área y perímetro de círculos, así como del volumen de esferas, los cuales son conceptos básicos pero esenciales en matemáticas y física.
Sin embargo, realizar estos cálculos de manera manual puede ser propenso a errores y consumir tiempo valioso. Por ello, se requiere desarrollar un programa en Python que permita al usuario calcular de forma rápida y precisa estas medidas geométricas mediante un menú interactivo.

### Fórmulas Matemáticas:
* Círculo:
Área: A = π × r²
Perímetro: P = 2 × π × r
* Esfera:
Volumen: V = (4/3) × π × r³

* donde La letra $r$ representa el Radio. Es la distancia más importante en cualquier figura circular o esférica.
* Se define como el segmento que une el centro de la figura con cualquier punto de su borde (o superficie, en el caso de la esfera).
* Si cambias el radio, cambias el tamaño de toda la figura de forma proporcional.Dato curioso: Si juntas dos radios en línea recta pasando por el centro, obtienes el Diámetro.
* 🥧 El número Pi ($\pi$)El símbolo $\pi$ es una constante matemática. No importa si el círculo es del tamaño de una moneda o del tamaño de un planeta, si divides su perímetro entre su diámetro, ¡siempre obtendrás el mismo número! Aproximadamente 3.14159.
* Es el "ingrediente secreto" que conecta las líneas rectas con las curvas.📏 Las Medidas FinalesÁrea ($A$): Representa la "superficie" o cuánto espacio plano ocupa el círculo.
* Imagina cuánta pintura necesitarías para rellenarlo.Perímetro ($P$): Es la longitud de la línea exterior.
* Si cortaras el círculo y estiraras el borde como una cuerda, esa medida sería el perímetro.Volumen ($V$): Esto ya es en 3D.Representa cuánto espacio ocupa la esfera o cuánta agua le cabría si fuera un balón hueco.



```python
# -----------------------------------
# problema de calculo geometrico
# programa realizado por jose solis
# fecha 13/04/2026
# -----------------------------------
import math

# -----------------------------------
def areacirculo():
    """Calcula el área de un círculo"""
    print("\n📐 CÁLCULO DEL ÁREA DEL CÍRCULO")
    print("Fórmula: Área = π × r²")
    radio = float(input("Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"✅ El área del círculo con radio {radio} es: {area:.2f}")

def perimetrocirculo():
    """Calcula el perímetro de un círculo"""
    print("\n📏 CÁLCULO DEL PERÍMETRO DEL CÍRCULO")
    print("Fórmula: Perímetro = 2 × π × r")
    radio = float(input("Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f}")

def volumenesfera():
    """Calcula el volumen de una esfera"""
    print("\n🔮 CÁLCULO DEL VOLUMEN DE LA ESFERA")
    print("Fórmula: Volumen = (4/3) × π × r³")
    radio = float(input("Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f}")

def menu():
    """Muestra el menú de opciones"""
    print("\n" + "="*50)
    print("🔵 MENÚ PRINCIPAL 🔵")
    print("="*50)
    print("1️⃣  Calcular área del círculo")
    print("2️⃣  Calcular perímetro del círculo")
    print("3️⃣  Calcular volumen de la esfera")
    print("4️⃣  Salir del programa")
    print("="*50)

def main():
    """Función principal del programa"""
    opcion = 0

    while opcion != 4:
        menu()
        opcion = int(input("\n👉 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta luego 🎓")
        else:
            print("\n❌ Opción no válida. Por favor selecciona una opción del 1 al 4.")

        if opcion != 4:
            input("\nPresiona Enter para continuar...")

# Ejecutar el programa
if __name__ == "__main__":
    main()

print("programa realizado por jose solis NL 51")
```
